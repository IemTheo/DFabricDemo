# Adjust to naming convention and data type

```python
# Declare the path to file
csv_path = "Files/csv/end_products.csv"

# Read the csv file
df_end_products = spark.read.csv(csv_path,header=True, inferSchema='true')

# Changing column names
df_end_products = df_end_products.withColumnRenamed("SKUCode","sku_code")\
    .withColumnRenamed("SKUDescription","sku_description")\
    .withColumnRenamed("CategoryCode","category_code")\
    .withColumnRenamed("CategroryDescription","categrory_description")\
    .withColumnRenamed("SubcategoryCode","subcategory_code")\
    .withColumnRenamed("SubcategoryDescription","subcategory_description")\
    .withColumnRenamed("Flavour","flavour")\
    .withColumnRenamed("Volume","volume_liter")\
    .withColumnRenamed("Price","price_eur")\
    .withColumnRenamed("ShelfLife","shelf_life")
    
```

```python
df_end_products.printSchema()
```

## SQL filtering

### LIKE

```python
# Filter where sku_code starts with 2
df_end_products.filter(df_end_products.sku_code.like("2%")).show()
```

```python
# Filter where sku_code starts with 2
df_end_products.filter(df_end_products.sku_code.like("%9")).show()
```

### WHERE

```python
# Show only banana flavored ice creams
df_end_products.where(df_end_products.flavour == 'Banana').show()
```

### Simple aggregates

```python
# Counting rows in the group. How many sizes are available for each flavour?
df_end_products.groupBy('flavour').count().show()
```

```python
# Maximum sales price for each flavor
df_end_products.groupBy('flavour')\
    .max('price_eur')\
    .withColumnRenamed('max(price_eur)','max_price_eur')\
    .sort('max_price_eur',ascending=False)\
    .show()
```

```python
# Maximum sales price for each flavor
from pyspark.sql.functions import max

df_end_products.groupBy('flavour')\
    .agg(max('price_eur').alias('max_price_eur'))\
    .sort('max_price_eur',ascending=False)\
    .show()
```

### HAVING

```python
# Maximum sales price for each flavor
from pyspark.sql.functions import avg, max, min, round, col

df_end_products.groupBy('flavour')\
    .agg(max('price_eur').alias('max_price_eur'), min('price_eur').alias('min_price_eur'), avg('price_eur').alias('avg_price_eur'))\
    .sort('max_price_eur',ascending=False)\
    .show()
```

```python
# Maximum sales price for each flavor
from pyspark.sql.functions import avg, max, min, round, col

df_end_products.groupBy('flavour')\
    .agg(max('price_eur').alias('max_price_eur'), min('price_eur').alias('min_price_eur'), round(avg('price_eur'),0).alias('avg_price_eur'))\
    .sort('max_price_eur',ascending=False)\
    .where(col('avg_price_eur') > 5)\
    .show()
```
